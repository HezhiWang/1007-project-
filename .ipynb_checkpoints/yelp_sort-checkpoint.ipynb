```python
def sort_trip(df):
    '''for museums/attractions with ratings higher than 3.5 and total number of reviews 
    more than 1000, sort by rating and total reviews descending.
    '''
    mask1 = df['rating']>=3.5
    mask2 = df['total_review']>=1000
    mask = pd.concat((mask1, mask2), axis=1)
    ind_sub = mask.all(axis=1)
    df_sub = df.ix[ind_sub]
    df_sorted = df_sub.sort_values(by=['rating','total_review'],ascending=[False,False])
    return df_sorted
```

```python
d = pd.concat([d1,d2],ignore_index=True)
d = d.drop_duplicates(subset=['name'],keep='first')
d = sort_trip(d).reset_index(drop=True)
```

```python
def print_to_rtf(df, filename):
    '''print sorted museums/attractions to rtf file'''
# replace - "-999" -> "NA"
    df = df.replace(to_replace= '-999', value='N.A.')
# open file
    rtf = open(filename+'.rtf', 'w')
# header info
    rtf.write(r'{\rtf1\ansi\ansicpg1252\deff0\deflang1033{\fonttbl{\f0\fswiss\fcharset0 Arial;}}')
    for i in range(df.shape[0]):
        rtf.write(r'\ \b {} \b0 \line \ \b Rating:\b0 {} \t \b Reviews:\b0 {} \line'.format(df['name'].iloc[i],
               df['rating'].iloc[i], df['total_review'].iloc[i]))
        rtf.write(r'\ \b Details: \b0 {} \line\ \b Description: \b0 {} \line'.format(df['detail'].iloc[i],
                                                                    df['description'].iloc[i]))
        rtf.write(r'\line')
    rtf.write(r'}\n\x00')
    rtf.close()
```

```python
def plot_review_density(df,filename):
    '''density plot of all reviews ''' 
    reviews = df[df['total_review']>=0]['total_review']
    mean = np.mean(reviews)
    std = np.std(reviews)
    median = np.median(reviews)
    ax1 = reviews.plot(kind='kde')
    ax1.set_xlim([0,max(reviews)])
    plt.xlabel('number of reviews')
    plt.ylabel('density')
    plt.title(filename+'_reviews_density_plot')
    plt.text(60000, 0.00010,'mean={:.2f} \n std={:.2f} \n median={:.2f}'.format(mean,std,median))
    plt.savefig(filename+'_reviews_density.png')
```

```python
plot_review_density(d1,'museum')
```

```python
def plot_rating_bar(df,filename):
    '''plot barplot of rating'''
    rating = df[df['rating']!= -999]['rating']
    mean = np.mean(rating)
    std = np.std(rating)
    ax = rating.value_counts().plot(kind='bar')
    counts = ax.get_yticks()  
    tot = sum(count)
    ax.set_yticklabels(['{:2.1f}%'.format(x/tot*100) for x in counts])
    plt.xlabel('rating')
    plt.ylabel('ratio')
    plt.text(0.5, 0.5,'mean={:.2f}, std={:.2f}'.format(mean,std))
    plt.title(filename+' ratings')
    plt.savefig(filename+'_ratings.png')
```

```python
#from reportlab.lib.pagesizes import letter
def hello(c):
    from reportlab.lib.units import inch
    c.translate(inch,inch)
    c.setFont("Helvetica", 14)
    c.setStrokeColorRGB(0.2,0.5,0.3)
    c.setFillColorRGB(1,0,1)
     # draw some lines
    c.line(0,0,0,1.7*inch)
    c.line(0,0,1*inch,0)
     # draw a rectangle
    c.rect(0.2*inch,0.2*inch,1*inch,1.5*inch, fill=1)
    # make text go straight up
    c.rotate(90)
    # change color
    c.setFillColorRGB(0,0,0.77)
    # say hello (note after rotate the y coord needs to be negative!)
    c.drawString(0.3*inch, -inch, "Hello World")
    c.drawString(100,100,"Hello World")
c = canvas.Canvas('myfile.pdf')
hello(c)
c.showPage()
c.save()
```
